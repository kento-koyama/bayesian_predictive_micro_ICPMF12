```r
# Growth model example, with stochastic growth model ("pure birth process").
# Generate some test data, with given parameters:
# growth rate mutrue, and zero lag, and N0:
mu.true <- 0.01   # check the posterior distribution of mu in the end!
N0 <- 10  # initial bacteria count

# generate the stochastic process for making data:
t<- numeric()
N <-  numeric()
t[1] <- 0
N[1] <- N0
for(i in 2:10000){
t[i] <- t[i-1]+rexp(1,mu.true*N[i-1])  # increment times (births) are exponentially distributed
N[i] <- N[i-1]+1 # counter for the cell count.
}

time <- c(0,50,100,300,600) # observation times
T <- length(time) # number of observation times
nobs <- numeric()  # observed cell counts at the given times
nobs[1]<-N0
nobs[2]<-N[sum(t<time[2])]
nobs[3]<-N[sum(t<time[3])]
nobs[4]<-N[sum(t<time[4])]
nobs[5]<-N[sum(t<time[5])]

# or simply set some data values:
#nobs <- c(10,16,28,178,3098)
#time <- c(0,50,100,300,600)
#T <- length(time)

print("Observed counts were:")
print(nobs)
plot(time,nobs,pch=16,main="Observed counts N(t)")
############################################
## MODEL 1:  this assumes we have observed the true bacteria counts
## consecutively at the observation times from the same growth process.
## BUGS model code to be contained in the file "growthmodel_1.txt" given to BUGS among inputs:
growthmodel_same_process <- function(){
for(t in 2:T){
# data: observed counts with no errors,
# observed increments from the same process, step-by-step
increment[t] <- nobs[t]-nobs[t-1]
increment[t] ~ dnegbin(prob[t],nobs[t-1])  # increments have negbinomial-distribution
prob[t] <- exp(-mu*max(time[t]-tnull[t],0) )
tnull[t] <- time[t-1]*step(time[t-1]-lag) +  lag*(1-step(time[t-1]-lag))
}
prob[1] <- 1-exp(0)
increment[1] ~ dpois(1) # dummy variable, not observed
# priors & constants:
mu ~ dunif(0,0.2)
lag <- 0  # lag is defined here as constant within the model code
}
#############################################
file.name <- "growthmodel_1.txt"
write.model(growthmodel_same_process,file.name)

data <- list("nobs","T","time")  # data setup
inits <- function(){list(mu=0.05)}  # initial values setup
parameters <- c("mu")  # parameters to get from results as MCMC sample. (mu = growth rate)

# Call BUGS to run the MCMC:
results1 <- bugs(data,inits,parameters,model.file="growthmodel_1.txt",
n.chains=1,n.burnin=1000,n.iter=11000,DIC=FALSE,codaPkg=FALSE,debug=FALSE)

################################################
## MODEL 2:  this assumes we have observed the true bacteria counts
## at the observation times from separate growth processes independently.
## BUGS model code to be contained in the file "growthmodel_2.txt" given to BUGS among inputs:
growthmodel_different_process <- function(){

for(t in 2:T){
  # data: observed counts with no errors,
  # observed increment from each process, time[t] vs the first time[1]
increment[t] <- nobs[t]-nobs[1]
increment[t] ~ dnegbin(prob[t],nobs[1]) # increments have negbinomial-distribution
prob[t] <- exp(-mu*max(time[t]-lag,0) )
}
prob[1] <- 1-exp(0)
increment[1] ~ dpois(1) # dummy variable, not observed
# priors & constants:
mu ~ dunif(0,0.2)
lag <- 0  # lag is defined here as constant within the model code
}
###############################################
file.name <- "growthmodel_2.txt"
write.model(growthmodel_different_process,file.name)

# Call BUGS to run the MCMC:
# data setup and initial values set up: the same as defined previously
results2 <- bugs(data,inits,parameters,model.file="growthmodel_2.txt",
n.chains=1,n.burnin=1000,n.iter=11000,DIC=FALSE,codaPkg=FALSE,debug=FALSE)

# plot the posterior density of mu, based on model1:
attach.bugs(results1)
plot(density(mu),col="blue",xlim=c(0.005,0.02),xlab="mu",main="Posterior density of mu")
quantile(mu,c(0.025,0.975))
# plot the posterior density of mu, based on model2:
attach.bugs(results2)
lines(density(mu),col="red",xlab="mu",main="Posterior density of mu")
points(0.01,0,pch=16)
legend("topright",legend=c("As single process data","As separate processes data"),text.col=c("blue","red"))
```

```r
##########  Next model with also lag-parameter
# Generate some test data, with given parameters:
# growth rate mutrue, and zero lag, and N0:
mutrue <- 0.011  # check the posterior distribution of mu in the end!
N0 <- 10
lagtrue <- 30
t<- numeric()
N <-  numeric()
t[1] <- 0
N[1] <- N0
t[2] <- lagtrue
N[2] <- N0
for(i in 3:1600000){
t[i] <- t[i-1]+rexp(1,mutrue*N[i-1])
N[i] <- N[i-1]+1
}

time <- c(0,50,100,300,600) # observation times
T <- length(time)
nobs <- numeric()
nobs[1]<-N0
nobs[2]<-N[sum(t<time[2])]
nobs[3]<-N[sum(t<time[3])]
nobs[4]<-N[sum(t<time[4])]
nobs[5]<-N[sum(t<time[5])]

##############################################
growthmodel_same_process_obs_error <- function(){
# data: observed counts WITH errors,
# observed counts from the same process, step-by-step
for(t in 2:T){
nobser[t] ~ dpois(lambda[t]) # observation as Poisson variable
lambda[t] <- ntrue[t]*0.3  # assume sample of 30% of assay volume for observations
# (assuming the observation sample is returned back to the population, i.e. no loss)
ntruehat[t] <- nobser[t]/0.2 # simple point estimate for true count
ntrue[t] <- ntrue[t-1]+increment[t]
increment[t] ~ dnegbin(prob[t],ntrue[t-1])
prob[t] <- exp(-mu*max(time[t]-tnull[t],0) )
tnull[t] <- time[t-1]*step(time[t-1]-lag) +  lag*(1-step(time[t-1]-lag))
}
nobser[1] ~ dpois(lambda[1])
lambda[1] <- ntrue[1]*0.3
ntruehat[1] <- nobser[1]/0.3
prob[1] <- 1-exp(0) # dummy
increment[1] <- 0 # dummy variable, not observed
# prior
mu ~ dunif(0,0.2)
ntrue[1] ~ dcat(p[1:100])
for(i in 1:100){p[i] <- 1/100} # discrete uniform distribution from 1 to 100
invmean <- 1/(10)
lag ~ dlnorm(0,1)   # prior
one <- 1
one ~ dbern(pconstraint)   # to constrain lag < time[4]
pconstraint <- step(time[4]-lag)*step(lag)
}
###################################################

file.name <- "growthmodel_12.txt"
write.model(growthmodel_same_process_obs_error,file.name)

ntrue_orig <- numeric()
ntrue_orig[1:T] <- nobs[1:T]
nobser <- numeric()
  # previously nobs was assumed exact observation of counts
  # now it is assumed to be the result of sampling 20% of the assay volume
  nobser[1:T] <- rpois(T,ntrue_orig[1:T]*0.3) # this is the new data
  print(paste(c("Ntrue:",ntrue_orig)))
  print(paste(c("Nobserved:",nobser)))

data <- list("nobser","T","time")
inits <- function(){list(mu=0.05,lag=time[2]/2,increment=c(NA,6,12,150,2920))}
parameters <- c("mu","ntrue","lag")  # mu = growth rate, ntrue = true counts
results12 <- bugs(data,inits,parameters,model.file="growthmodel_12.txt",
n.chains=1,n.burnin=5000,n.iter=55000,DIC=FALSE,codaPkg=FALSE,debug=FALSE)
attach.bugs(results12)

pairs(cbind(mu,ntrue[,1],lag),labels=c(expression(mu),expression(N[0]),"lag"),pch=16,cex=0.5,col="forestgreen")
#plot(density(mu),col="blue",xlim=c(0.005,0.02),main="Posterior of mu")
#hist(ntrue[,1],50,xlab=expression(N[0]==N[t[1]]),main="Posterior of N0",probability=TRUE)
#plot(ntrue[,1],mu,pch=16,cex=0.8,xlab="N0",ylab="mu",col="forestgreen")
```

```r
# Censored data example, (left-censored data), lognormal model.
# With artificial data for 'concentration measurements' with some LOQ,
# lots of censored values:
LOQ <- 3
c <- LOQ # set censoring limit (limit of quantification)
n <- 50 # number of measurements (data sample size)
# y <- rnorm(n,mean=1,sd=2) # create data randomly, or use the data below
# m <- sum(y>c) # number of accurate values (not censored)
# y <- c(rep(NA,n-m),y[y>c]) # vector of accurate values and NA for censored
y <- numeric()
y[12:50] <- NA   # there were 39 values <LOQ
y[1:11] <- c(3.01, 3.14, 3.17, 3.23, 3.37, 3.37, 3.44, 3.51, 3.87, 4.92, 5.89) # values >LOQ
m <- 11 # number of accurate values (not censored)
n <- length(y) # number of all data values

sig.up <- 10 # upper limit for sigma prior, set as needed
mu.lo <- -10 # lower limit for mu prior, set as needed
mu.up <- 10 # upper limit for sigma prior, set as needed
data<-list("y","c","n","m","sig.up","mu.lo","mu.up") # data setup
inits<-function(){list(mu=0,sig=1)}   # initial values setup
parameters<-c("mu","sig") # parameters to get from results as MCMC sample

##########################################
# BUGS model code to be contained in the file "cenmodel.txt" given to BUGS among inputs:
censoreddatamodel <- function(){
for(i in 1:m){ y[i] ~ dnorm(mu,tau) } # model for accurate data points
for(i in m+1:n){  # model for censored data points
  #y[i] ~ dnorm(mu,tau)  # C(,c)  # this syntax for censoring is normally used!
  one[i] <- 1    # this is alternative way to add the same information of censored data.
  one[i] ~ dbern(pr[i])  # probability for the observed event: "it was left-censored"
  pr[i] <- phi((c-mu)*sqrt(tau)) # the probability of censoring equals the cumulative probability at c=LOQ
  }
mu ~ dunif(mu.lo,mu.up) # prior for mu
tau <- pow(sig,-2) # conversion from standard deviation to precision parameter
sig ~ dunif(0,sig.up) # prior for sigma
}
###########################################
file.name <- "cenmodel.txt"  # BUGS model code is in this file
write.model(censoreddatamodel,file.name)

# Call BUGS to run the MCMC:
mymodelresult <- bugs(data,inits,parameters,model.file="cenmodel.txt",
n.iter=10500,n.burnin=500,n.chains=1)
attach.bugs(mymodelresult)

# plot the 2D posterior distribution (MCMC sample) of the parameters:
plot(mu,sig,pch=16,cex=0.4,
ylab="sigma",xlab="mu",main="Posterior distribution of mu & sigma",
col=densCols(mu,sig,colramp=colorRampPalette(c("darkgreen","yellow","red","magenta","black"))))
# show the true parameter values:
points(1,2,pch=16,col="blue",cex=2)

x <- seq(0,50,length=100)
plot(x,dlnorm(x,mu[1],sig[1]),type="l",col="cyan",
main="Uncertainty of the lognormal(mu,sigma) -distribution",xlab="Concentration",ylab="Density")
index <- round(seq(1,10000,length=100))
for(i in 1:100){
lines(x,dlnorm(x,mu[index[i]],sig[index[i]]),col="cyan")
}

# more reading in chapter 6 of the book: https://doi.org/10.1201/9780429083525

```

```r
### Optional Example:  dose-response estimation for 3 campylobacter strains
### Model: multilevel (hierarchical) model for dose-response parameters

# data from the publication (2018):
# https://doi.org/10.1016/j.epidem.2018.02.001

typenames=c(
"A3249","A3249","A3249","A3249","A3249","A3249","A3249",
"81-176","81-176","81-176",
"CG8421","CG8421","CG8421",
"81-176","81-176","81-176","81-176","81-176",
"CG8421","CG8421","CG8421","CG8421","CG8421","CG8421")

#log10dose, infcases, num, year.of.reference
DATA <- matrix(c(
2.90309, 5, 10, 2018,
3.90309, 6, 10, 2018,
4.954243, 11, 13, 2018,
5.90309, 8, 11, 2018,
6.0, 15, 19, 2018,
8.0, 5, 5, 2018,
8.0, 4, 4, 2018,
6.0, 7, 7, 2018,
8.0, 10, 10, 2018,
9.30103, 22, 22, 2018,
5.986772, 8, 8, 2018,
4.924279, 7, 7, 2018,
4.732394, 8, 8, 2018,
5.0, 5, 5, 2018,
7.0, 5, 5, 2018,
9.0, 36, 36, 2018,
9.0, 6, 8, 2018,
9.0, 7, 7, 2018,
5.69897, 15, 15, 2018,
5.69897, 8, 8, 2018,
4.278754, 6, 6, 2018,
4.30103, 7, 7, 2018,
5.146128, 4, 4, 2018,
5.255273, 13, 13, 2018),4,24)
DATA <- t(DATA)

##############################################
# MODEL: Beta-binomial dose-response model
drmodel <- function(){
for(k in 1:N){ # for each experiment in the data
infcases[k] ~ dbin(pinf[k],num[k])  # P(inf) depends on strain
d[k] <- pow(10,log10dose[k])  # dose is assumed to be known
pinf[k] <- 1 - exp(loggam(a[typeindex[k]] + b[typeindex[k]]) + loggam(b[typeindex[k]] + d[k]) - loggam(b[typeindex[k]]) - loggam(a[typeindex[k]] + b[typeindex[k]] + d[k]))
}
for(j in 1:J){ # for each strain, strain specific parameters
a[j] <- exp(loga[j])
b[j] <- exp(logb[j])
loga[j] ~ dnorm(loga0,tau.a)  # variation of strains around average parameters
logb[j] ~ dnorm(logb0,tau.b)
}
loga0 ~ dnorm(0,0.01) # prior for average parameters
logb0 ~ dnorm(0,0.01)
tau.a ~ dgamma(1,1)  # prior for strain variation
tau.b ~ dgamma(1,1)
}
##############################################
file.name <- "drmodel.txt"
write.model(drmodel,file.name)

N <- length(typenames) # number of experiments in the data
log10dose <- DATA[,1]
infcases <- DATA[,2]
num <- DATA[,3]
types <- unique(typenames)
J <- length(types)  # number of different strains
typeindex <- numeric(length(typenames))  # create numerical indexing of strain types
for(i in 1:length(typenames)){typeindex[typenames==types[i]]<-i}

data <- c("log10dose","infcases","num","typeindex","N","J")  # data setup
inits <- function(){list(loga0=0,logb0=0,
loga=rep(0,max(typeindex)),logb=rep(0,max(typeindex)),
tau.a=1,tau.b=1)}   # initial values setup
parameters= c("loga0","logb0","loga","logb","tau.a","tau.b") # parameters to get from results as MCMC sample
# call BUGS to run the MCMC:
result <- bugs(data,inits,parameters,model.file="drmodel.txt",
    n.burnin=1000,n.iter=16000,n.chains=1,DIC=FALSE)
attach.bugs(result)

# plot the posterior distributions for the DR-parameters for 3 strains:
plot(loga[,1]/log(10),logb[,1]/log(10),pch=16,col="red",cex=0.2,xlim=c(-4,3),ylim=c(-6,4),xlab="log10(a)",ylab="log10(b)",main="Posteriors of DR-parameters for 3 strains")
points(loga[,2]/log(10),logb[,2]/log(10),pch=16,col="blue",cex=0.2)
points(loga[,3]/log(10),logb[,3]/log(10),pch=16,col="green",cex=0.2)

# plot the uncertainty plots for the DR-functions for each strain,
# and the point estimates from data samples (#inf/#num.exposed)
a0<-exp(loga0) # transform from log-values
b0<-exp(logb0)
a <- exp(loga)
b <- exp(logb)
log10drange <- seq(0,9,length=100) # create a plotting range for dose
drange <- 10^log10drange
s <- round(seq(1,15000,length=200)) # get 200 samples from the full MCMC sample
par(mfrow=c(2,2))

# plot DR for strain1
plot(log10dose[typeindex==1],infcases[typeindex==1]/num[typeindex==1],pch=16,cex=1.5,col="black",xlim=c(0,9),ylim=c(0,1),main="Strain 1",xlab="log10(dose)",ylab="P(inf)")
for(i in 1:length(s)){
lines(log10drange,
1 - exp(lgamma(a[s[i],1] + b[s[i],1]) + lgamma(b[s[i],1] + drange) - lgamma(b[s[i],1]) - lgamma(a[s[i],1] + b[s[i],1] + drange)),col="red")
}
points(log10dose[typeindex==1],infcases[typeindex==1]/num[typeindex==1],pch=16,cex=1.5,col="black")
# plot DR for strain2
plot(log10dose[typeindex==2],infcases[typeindex==2]/num[typeindex==2],pch=16,cex=1.5,col="black",xlim=c(0,9),ylim=c(0,1),main="Strain 2",xlab="log10(dose)",ylab="P(inf)")
for(i in 1:length(s)){
lines(log10drange,
1 - exp(lgamma(a[s[i],2] + b[s[i],2]) + lgamma(b[s[i],2] + drange) - lgamma(b[s[i],2]) - lgamma(a[s[i],2] + b[s[i],2] + drange)),col="blue")
}
points(log10dose[typeindex==2],infcases[typeindex==2]/num[typeindex==2],pch=16,cex=1.5,col="black")
# plot DR for strain3
plot(log10dose[typeindex==3],infcases[typeindex==3]/num[typeindex==3],pch=16,cex=1.5,col="black",xlim=c(0,9),ylim=c(0,1),main="Strain 3",xlab="log10(dose)",ylab="P(inf)")
for(i in 1:length(s)){
lines(log10drange,
1 - exp(lgamma(a[s[i],3] + b[s[i],3]) + lgamma(b[s[i],3] + drange) - lgamma(b[s[i],3]) - lgamma(a[s[i],3] + b[s[i],3] + drange)),col="green")
}
points(log10dose[typeindex==3],infcases[typeindex==3]/num[typeindex==3],pch=16,cex=1.5,col="black")

# plot DR for a general strain, based on hyper parameters ('parent parameters' for all strains)
plot(log10dose,infcases/num,pch=16,cex=1.5,col="black",xlim=c(0,9),ylim=c(0,1),xlab="log10(dose)",ylab="P(inf)",main="General")
for(i in 1:length(s)){
lines(log10drange,
1 - exp(lgamma(a0[s[i]] + b0[s[i]]) + lgamma(b0[s[i]] + drange) - lgamma(b0[s[i]]) - lgamma(a0[s[i]] + b0[s[i]] + drange)),col="gray")
}
points(log10dose,infcases/num,pch=16,cex=1.5,col="black")
```
